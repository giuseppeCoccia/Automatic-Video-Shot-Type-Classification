```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
def read_csv(filename):
    losses, train_accs, valid_accs = [], [], []
    
    with open(filename, newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter=';', quotechar='|')
        next(reader, None)
        for row in reader:
            losses.append(float(row[0]))
            train_accs.append(float(row[1]))
            valid_accs.append(float(row[2]))
        
    return losses, train_accs, valid_accs

def plot_loss(losses, title):
    plt.figure(figsize=(23, 5))
    plt.grid()
    plt.xticks(range(0, len(losses), 5))
    plt.title(title)
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.xlim(0, len(losses)-1)
    plt.plot(range(len(losses)), losses)
    
def plot_train_valid_accuracy(train_accs, valid_accs, title):
    plt.figure(figsize=(23, 5))
    plt.grid()
    plt.xticks(range(0, len(losses), 5))
    plt.title(title)
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.xlim(0, len(losses)-1)
    plt.plot(range(len(train_accs)), train_accs, label="Train_accuracy")
    plt.plot(range(len(valid_accs)), valid_accs, label="Validation_accuracy")
    plt.legend()
    plt.show()
```

# 10. Accuracy after reducing training size

```python
def plot_train_valid_accuracies(train_accs, valid_accs, title):
    plt.figure(figsize=(23, 12))
    plt.grid()
    plt.xticks(range(0, len(train_accs[0]), 5))
    plt.title(title)
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.xlim(0, len(train_accs[0])-1)
    
    colors = ['r', 'g', 'b', 'y', 'c']
    train_size = ['933', '651', '402', '102', '30']
    for train_acc, valid_acc, color, tr_size in zip(train_accs, valid_accs, colors, train_size):
        plt.plot(range(len(train_acc)), train_acc, linestyle="dashed", c=color, label="Train_accuracy"+tr_size)
        plt.plot(range(len(valid_acc)), valid_acc,  c=color, label="Validation_accuracy"+tr_size)
    
    plt.legend()
    plt.show()
```

```python
_, train_accs_933, valid_accs_933 = read_csv("resnet_933.csv")
_, train_accs_651, valid_accs_651 = read_csv("resnet_651.csv")
_, train_accs_402, valid_accs_402 = read_csv("resnet_402.csv")
_, train_accs_102, valid_accs_102 = read_csv("resnet_102.csv")
_, train_accs_30, valid_accs_30 = read_csv("resnet_30.csv")
plot_train_valid_accuracies([train_accs_933[:60], train_accs_651[:60], train_accs_402[:60], train_accs_102[:60], train_accs_30[:60]], 
                            [valid_accs_933[:60], valid_accs_651[:60], valid_accs_402[:60], valid_accs_102[:60], valid_accs_30[:60]], "Accuracy on ResNet152 with learning_rate=0.001")
```
