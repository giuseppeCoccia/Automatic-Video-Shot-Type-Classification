```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
def read_csv(filename):
    losses, train_accs, valid_accs = [], [], []
    
    with open(filename, newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter=';', quotechar='|')
        next(reader, None)
        for row in reader:
            losses.append(float(row[0]))
            train_accs.append(float(row[1]))
            valid_accs.append(float(row[2]))
        
    return losses, train_accs, valid_accs

def plot_loss(losses, title):
    plt.figure(figsize=(23, 5))
    plt.grid()
    plt.xticks(range(len(losses)))
    plt.title(title)
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.xlim(0, len(losses)-1)
    plt.plot(range(len(losses)), losses)
    
def plot_train_valid_accuracy(train_accs, valid_accs, title):
    plt.figure(figsize=(23, 5))
    plt.grid()
    plt.xticks(range(len(losses)))
    plt.title(title)
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.xlim(0, len(losses)-1)
    plt.plot(range(len(train_accs)), train_accs, label="Train_accuracy")
    plt.plot(range(len(valid_accs)), valid_accs, label="Validation_accuracy")
    plt.legend()
```

# 1. ResNet

## 1.1 Tanh features transformation

```python
losses, train_accs, valid_accs = read_csv("out_tanh_0.01lr.csv")
plot_loss(losses, "Loss on ResNet with learning_rate=0.01 and tanh as features transformation")
plot_train_valid_accuracy(train_accs, valid_accs, "Accuracy on ResNet with learning_rate=0.01 and tanh as features transformation")
```

```python
losses, train_accs, valid_accs = read_csv("out_tanh_0.001lr.csv")
plot_loss(losses, "Loss on ResNet with learning_rate=0.001 and tanh as features transformation")
plot_train_valid_accuracy(train_accs, valid_accs, "Accuracy on ResNet with learning_rate=0.001 and tanh as features transformation")
```
